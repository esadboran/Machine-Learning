# 1. Introduction

K-Nearest Neighbor algorithm was utilized in this experiment to solve the classification and regression problems. The K-Nearest Neighbors algorithm calculates the likelihood that a data point will join one group or another based on which group the data points closest to it are members of. The input for classification and regression consists of a data set's k closest training instances.  This assignment is divided into two sections. A KNN-classification experiment is included in the first section, and a KNN-regression experiment is included in the second.

#    4.1 Part I: Personality Classification

```python
# We got the data with the pandas library
data = pd.read_csv("subset_16P.csv")
data.head()
```

```python
# We converted the dataframe to np.array to process easier on the data using the numpy library.
arr_data = np.array(data)
```

```python
# We have generated the shuff() method to prevent data in the dataset from being collected in certain places.
# This method randomly replaces the elements in the dataset.
shuff_data = shuff(arr_data)
```

```python
# There are 16 different Personality attribute in type_labels. This will be used to calculate recall and precise.
type_labels = pd.unique(data.Personality)
type_labels
```

## 4.1.1 Data preprocessing and separation

```python
# 60 questions and their answers.
X = np.array(shuff_data[:,1:-1])
```

```python
# The labels of data.
y = np.array(shuff_data[:,-1])
```

```python
# We used normalize() function to apply min-max scaling to the dataset.
X_normalized = normalize(X)
```

## 4.1.2 Unweighted KNN Classification

### 4.1.2.1 Data Without Normalization

```python
# The explanation of results_cv() function is at the background section. 
# Accuracy, precision, and recall ratios of unweighted and non-normalized K-NN model.
unweighted_without_norm = results_cv(X,y,model_type="Classification",weights=False,normalized = False,calculations="metric")
unweighted_without_norm # unweighted_without_norm = [accuracy table , precision table , recall table]
```

### 4.1.2.2 Data With Normalization

```python
# Accuracy, precision, and recall ratios of unweighted and normalized K-NN model.
unweighted_with_norm = results_cv(X,y,model_type="Classification",weights=False,normalized = True,calculations="metric")
unweighted_with_norm # unweighted_with_norm = [accuracy table , precision table , recall table]
```

## 4.1.3 Weighted KNN Classification

### 4.1.3.1 Data Without Normalization

```python
# Accuracy, precision, and recall ratios of weighted and non-normalized K-NN model.
weighted_without_norm = results_cv(X,y,model_type="Classification",weights=True,normalized = False,calculations="metric")
weighted_without_norm # weighted_without_norm = [accuracy table , precision table , recall table]
```

### 4.1.3.2 Data With Normalization

```python
# Accuracy, precision, and recall ratios of weighted and normalized K-NN model.
weighted_with_norm = results_cv(X,y,model_type="Classification",weights=True,normalized = True,calculations="metric")
weighted_with_norm # weighted_with_norm = [accuracy table , precision table , recall table]
```

## 4.1.4 Performance Metric

### 4.1.4.1 Accuracy Analysis

```python
data_accuracy = [unweighted_with_norm[0], unweighted_without_norm[0], weighted_with_norm[0],weighted_without_norm[0]]
data_accuracy = pd.concat(data_accuracy,axis=1)
data_accuracy.columns = [x for x in range(1,21)]
data_accuracy.to_csv('data_accuracy.csv', index=False) 
data_accuracy # In the data_accuracy table,#1-5 ----> Unweighted without normalization for k = 1,3,5,7,9
                                           #6-10----> Unweighted with normalization    for k = 1,3,5,7,9
                                           #11-15---> Weighted with normalization      for k = 1,3,5,7,9
                                           #16-20---> Weighted without normalization   for k = 1,3,5,7,9
```

# 5 Error Analysis

## 5.1 Error Analysis for Part 1

 In Part 1, all for Accuracy, Precision and Recall ratios, 5-Fold Cross Validation gives the best ratio.
 For K values, when K = 9 , our model gives the best ratio. (9 Nearest Neighbor)
 Weighted KNN models gives better ratios then Unweighted KNN models.
 Non-normalized KNN models gives better ratios then Normalized KNN models.
 
 Weighted KNN models gives better ratios because in weighted models, we consider not only nearest neighbors. We also consider how close they are. Weighted KNN models work better when the dataset is large.
 
 Non-normalized KNN models gives better ratios because our data's ranges are not large.(between -3 and +3) Normalized KNN models work better when data's ranges are large.
 
 Time analysis:
 
 In KNN.fit_transform(), we calculated the distances just one time to decrease the execution time. Thanks to this, we can say that our predict code works in just a few seconds. Total execution time for 10.000 sized dataset is approximately 35 minutes for 4 models.(Weighted/Unweighted , Normalized/Non-Normalized)
 The total execution time decreases to approximately 1 minute if we shrink the data size to 1000. The reason of this is calculating the distances are working in O(n^2).

## 5.2 Error Analysis for Part 2

In Part 2 for both Heating Load and Cooling Load, 1-Fold Cross Validation gives the minimum mean absolute error(mea). For K values, when K = 5, our model gives the estimation rate.(5 Nearest Neighbor) The reason that K = 5, is because our data is small.(768 sample) If our dataset was larger, K value might be higher for best estimation rate.

Unweighted KNN Model gives better estimation ratio then Weighted KNN Model. The reason for this is our dataset is small.
Normalized KNN Model gives better estimatiın ratio then Non-normalized KNN Model. The reason for this is our data's ranges are large.

Time analysis:
